```python
# https://github.com/jcgregorio/httplib2/wiki/Examples-Python3
import urllib
import httplib2
import json
import pandas as pd
```

```python
#mbsgs9
#mbsgs10
#mbsgstc9
#mbsgstc10

http = httplib2.Http()

domain = 'https://mbs.kt.com:8080'
url = domain + '/doLogin.do'
body = {'loginId':'mbsgs10','loginPw':'mbslive45@', 'channel': 'on'}
headers = {'Content-type': 'application/x-www-form-urlencoded'}
response, content = http.request(url, 'POST', headers = headers, body = urllib.parse.urlencode(body))

headers = {'Cookie': response['set-cookie']}
```

```python
headers
```

```python
url = domain + '/chart/rangeInterestCh.do?lastTime=20170615164900&screenType=S&_=1495606400857'
response, content = http.request(url, 'GET', headers=headers)
```

```python
text = content.decode('utf-8')
data = json.loads(text)
```

```python
len(data['chList'])
```

```python
data['chList']
```

```python
temp = []
for item in data['chList']:
    
    row = []    
    for temprow in item['list']:
        listitem = {}
        listitem['time'] = item['time']
        listitem['prgm_id'] = temprow['prgm_id']
        # listitem['ch_color'] = temprow['ch_color']
        listitem['ch_no'] = temprow['ch_no']
        listitem['ch_nm'] = temprow['ch_nm']
        listitem['disp_ch_nm'] = temprow['disp_ch_nm']
        listitem['view_cnt'] = temprow['view_cnt']
        listitem['tot_cnt'] = item['viewCntOTS'] + item['viewCntOTV']
        
        print(listitem)
        #row.append(listitem)
        temp.append(listitem)

temp
```

```python
df_mbs = pd.DataFrame(temp)
df_mbs
```

# CSV 출력하기
