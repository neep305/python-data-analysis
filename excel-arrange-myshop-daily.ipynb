```python
import pandas as pd
```

```python
from datetime import datetime, timedelta
```

```python
today = datetime.today() - timedelta(days=1)
str_today = today.strftime('%Y%m%d')

#str_today = '20170722'
str_today
```

```python
excel_result = pd.read_excel('./raw_data/유형별시청형태_채널_GS MY SHOP_시간별_1d_' + str_today + '.xlsx', sheetname='유형별>시간별_시청가구상세테이블',index_col=None, header=2)
```

```python
index =  range(1, len(excel_result)+1)
```

```python
temp = []
for i in index:
    temp.append(i)
```

```python
rearranged_excel = pd.DataFrame(excel_result, columns={'조회시간':'time','시청가구':'cnt'})
```

```python
drop_nan = rearranged_excel.reset_index(temp,drop=True)
```

```python
drop_nan.rename(columns={'조회시간':'time'}, inplace=True)
```

```python
drop_nan.rename(columns={'시청가구':'cnt'}, inplace=True)
```

```python
df3 = drop_nan[drop_nan.time != '일평균']
```

```python
df3.index
```

```python
df3_new = pd.DataFrame(df3, columns=['cnt','time'])
```

```python
df3_new.iloc[:,1] = str_today + df3_new['time'].str.replace(r'(시|분)','')
#df3.loc['time'] = str_today + df3['time'].str.replace(r'(시|분)','')
```

```python
df3_new
```

```python
temp_srvc = []
for i in range(len(df3_new)):
    #print(i)
    temp_srvc.append('T');
```

```python
df3_new['srvc'] = temp_srvc
```
