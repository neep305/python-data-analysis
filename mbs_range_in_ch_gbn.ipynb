```python
temp = []
for item in data['prevChList']:
    row = []    
    for temprow in item['list']:
        listitem = {}
        listitem['time'] = item['time']
        listitem['prgm_id'] = temprow['prgm_id']
        listitem['ch_color'] = temprow['ch_color']
        listitem['prev_ch_no'] = temprow['prev_ch_no']
        listitem['prev_ch_nm'] = temprow['prev_ch_nm']
        listitem['prev_disp_ch_nm'] = temprow['prev_disp_ch_nm']
        listitem['prev_prgm_id'] = temprow['prev_prgm_id']
        listitem['view_cnt'] = temprow['view_cnt']
        print(listitem)
        #row.append(listitem)
        temp.append(listitem)

temp
```
