```python
import pandas as pd
```

```python
from datetime import datetime, timedelta
```

```python
today = datetime.today() - timedelta(days=1)
str_today = today.strftime('%Y%m%d')

str_today
```

```python
excel_result = pd.read_excel('./raw_data/유형별시청형태_채널_GS MY SHOP_시간별_1d_20170629.xlsx', sheetname='유형별>시간별_시청가구상세테이블',index_col=None, header=2)
```

```python
index =  range(1, len(excel_result)+1)
```

```python
temp = []
for i in index:
    temp.append(i)
    
temp
```
